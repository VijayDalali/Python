```python
# 21 Differentiate between local and global variables with an example.
x = 100   # Global variable

def my_function():
    y = 50   # Local variable
    print("Inside function, x (global):", x)
    print("Inside function, y (local):", y)

my_function()
print("Outside function, x (global):", x)

#Key Difference: Global variable (x) → accessible everywhere. Local variable (y) → exists only inside the function where it is defined.
```

```python
#22 Write a function that modifies a global variable inside it.

x = 9   # Global variable

def change_x(): #defines a function named change_x
    global x    # Tell Python we mean the global x
    x = x + 1   # Modify global variable x by adding 1 to its current value.
    print("Inside function:", x)

change_x()
print("Outside function:", x)
```

```python
# 23 Why does Python discourage using variable names like l (lowercase L), O (capital O), or I (capital i)?
#l looks like 1 (one).
#O looks like 0 (zero).
#I looks like 1 (one) or lowercase l.

# Bad (confusing)
l = 5
O = 10
I = l + O
print(I)   # Is this 15 or "101"? Hard to read

# Instead, use descriptive names like length, order, index, etc.
# Good (clear)
length = 5
order = 10
index = length + order
print(index)  # 15
```

```python
#26 Demonstrate variable shadowing with a function.

x = "Global"   # Global variable

def my_function():
    x = "local"   # Local variable (shadows the global 'x')
    print("Inside function, x =", x) #executed inside my_function(), it prints the value of the local x

my_function()
print("Outside function, x =", x) #executed outside my_function(), it prints the value of the global x
```

```python
# 28 Show an example where using a mutable variable (like a list) as a default function argument causes unexpected behavior.
def collect_numbers(num, numbers=None): #Default is None (immutable). None is safe because it doesn't carry the same mutation semantics as a list.
    if numbers is None:          # Tests whether the caller provided a list. If not, create a fresh list.
        numbers = []         # Creates a new list each call when no list was provided.
    numbers.append(num)      # Appends to this fresh list (or to the explicitly provided list).
    return numbers           # Returns the list (fresh each call unless the caller passed one).

print(collect_numbers(1))   # [1]
print(collect_numbers(2))   # [2]   (fresh list each time)
print(collect_numbers(3))   # [3]


```
